```python
import pyautogui as p

# 해상도 확인
size = p.size()
print(size)
```

```python
# 마우스 조작

# 마우스 좌표 가져오기
position = p.position()
print (position)
print(position.x,position.y)
```

```python
# 마우스 이동 : moveTo(x,y,움직이는 시간) - 절대좌표
p.moveTo(100,100,0.5)
p.moveTo(200,200,0.5)
p.moveTo(300,300,0.5)
```

```python
# 마우스 이동 : moveTo(x,y,움직이는 시간),moveRel() - 상대좌표 
p.moveTo(100,100,0.5)
p.move(300,300,0.5)
print(p.position())
p.move(100,100,0.25)
```

```python
# 마우스 클릭
p.click()

# p.click(1970,14) : 해당좌표에서 클릭
```

```python
p.moveTo(x=1690, y=78)
p.drag(150,0,0.25)
```

```python
# dragTo() : 절대 위치로 drag

p.moveTo(x=1690, y=78)
p.dragTo(150,0,0.25)
```

```python
# scroll() : 양수 = 위, 음수 = 아래

p.moveTo(x=1690, y=90)
p.scroll(-100)
```

```python
# 스크린샷
# p.screenshot("파일명",region=()) : 특정위치에 대한 스크린 샷

# 전체화면
p.screenshot("./screen.png")
```

```python
p.screenshot("./file.png",region=(0,108,73,53))

p.sleep
file_location = p.locateOnScreen("./file.png")
print(file_location)
```

```python
file_location = p.locateOnScreen("./file.png",confidence=0.9)
p.click(file_location)
```

```python
# locateAllOnScreen()
p.sleep(1)
p.screenshot("./checkbox.png",region=(1688,337,15,15))

```

```python
for i in p.locateAllOnScreen("./checkbox.png"):
    print(i)
    p.click(i)
```

```python
# 윈도우
window = p.getActiveWindow()
print(window)
print(window.title)
print(window.top,window.left,window.width,window.height)
```

```python
for w in p.getAllWindows():
    print(w)
```

```python
# 특정창 가져오기
w = p.getWindowsWithTitle("제목 없음")
print(w)
```

```python
if not w.isActive():
    w.activate()
```

```python
if not w.isMaximized():
    w.maximize()
```

```python
if not w.isMinimized():
    w.minimize()
```

```python
# 키보드

w = p.getWindowsWithTitle("제목 없음")[0]
w.activate()

p.write("write")
```

```python
w = p.getWindowsWithTitle("제목 없음")[0]
w.activate()
p.write("pyautogui", interval=0.25)
```

```python
import pyperclip

w = p.getWindowsWithTitle("제목 없음")[0]
w.activate()
pyperclip.copy("안녕하세요") # 클립보드
p.hotkey("ctrl", "v") # ctrl +V
```

```python
def hangul_write(text):
    pyperclip.copy(text)
    p.hotkey("ctrl", "v")
```

```python
w = p.getWindowsWithTitle("제목 없음")[0]
w.activate()
hangul_write("반갑습니다")
hangul_write("hello")
```

```python
w = p.getWindowsWithTitle("제목 없음")[0]
w.activate()
#ctrl+A 키보드 누르기
p.keyDown("ctrl")
p.keyDown("a")
#ctrl+a 키보드에서 손때기 
p.keyUp("A")
p.keyUp("ctrl")
```
