#### 화면 조작
    -   find_element(By.ID)
    -   find_element(By.CLASS_NAME)
    -   find_element(By.XPATH)
    -   find_element(By.CSS_SELECTOR)

    -   click() : 클릭
    -   send_keys() : 키입력

```python
from chrome import set_chrome_driver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

driver = set_chrome_driver()

driver.get("https://www.naver.com")

# 검색창에 검색어 입력 후 엔터
# 검색 요소 찾기
element = driver.find_element(By.ID,"query")
# print(element)
element.send_keys("아이폰")
element.send_keys(Keys.ENTER)

# 검색 후 페이지 소스 가져오기
print(driver.page_source)


time.sleep(3)

driver.quit()
```

```python
from chrome import set_chrome_driver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

driver = set_chrome_driver()

driver.get("https://ecos.bok.or.kr/#/")
time.sleep(2)

# 100대 지표 클릭
button = driver.find_element(By.CSS_SELECTOR,"#root > div.wrap-body > div > div.main-wrap.clearfix > div.main-row3.clearfix > div.main-left > div.main-icon-menu > ul > li:nth-child(1) > a")
button.send_keys(Keys.ENTER)
time.sleep(2)

excel = driver.find_element(By.CSS_SELECTOR,"div.exelDown.partition-right > button")
excel.send_keys(Keys.ENTER)



# 검색 후 페이지 소스 가져오기
print(driver.page_source)


time.sleep(3)

driver.quit()
```
